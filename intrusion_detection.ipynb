```python
cols = ['logged_in',
 'count',
 'serror_rate',
 'srv_serror_rate',
 'rerror_rate',
 'srv_rerror_rate',
 'same_srv_rate',
 'dst_host_count',
 'dst_host_srv_count',
 'dst_host_same_srv_rate',
 'dst_host_diff_srv_rate',
 'dst_host_serror_rate',
 'dst_host_srv_serror_rate',
 'dst_host_rerror_rate',
 'dst_host_srv_rerror_rate',
 'class']

def data_cleaning(dataset, columns):
    dataset = dataset.filter(columns)
    dataset['class'] = dataset['class'].replace({'normal':0, 'anomaly':1})
    return dataset
```

```python
training_data.head()
```

```python
cleaned_training = data_cleaning(training_data, cols)
# cleaned_test = data_cleaning(test_data, cols)
```

```python
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split

X = cleaned_training.drop(labels='class', axis=1)
y = cleaned_training['class']

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.4,random_state =123)


clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(X_train, X_test, y_train, y_test)

print(models)
```
